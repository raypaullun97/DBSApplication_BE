```python
#extracting street name and postal code
list_postal_code= []
list_street= []
all_address = client_profile["Address"].tolist()
for each in all_address:
    start_digit_postal_code = each.find('Singapore') + 10
    postal_code = each[start_digit_postal_code:] 
    street = each[10:each.find('#')] 
    list_street.append(street)
    list_postal_code.append(int(postal_code))
client_profile['PostalCode'] = list_postal_code
client_profile['StreetName'] = list_street
client_profile
```

```python
#extracting street name and postal code
list_networth= []
all_income = client_profile["MonthlyIncome"].tolist()
for each in all_income:
    if(each >= 8000):
        list_networth.append('yes')
    else:
        list_networth.append('no')

client_profile['HighNetWorthIndividual'] = list_networth
client_profile
```
